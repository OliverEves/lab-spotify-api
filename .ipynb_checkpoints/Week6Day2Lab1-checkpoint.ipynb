```python
import spotipy
import json
from spotipy.oauth2 import SpotifyClientCredentials

sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(client_id= Client_ID,
                                                           client_secret= Client_Secret)) #connects to spotify api
 
```

```python
def search_song(title, artist):
    try:
    
        song = sp.search(q='track:' + title + ' artist:' + artist, limit=1)
        return song["tracks"]["items"][0]["id"]
    except:
        return 'unknown'
```

```python
def get_audio_features(list_of_songs):
    
    d = {'title':[],
        'artist':[],
             'danceability': [],
             'energy': [],
             'key': [],
             'loudness': [],
             'mode': [],
             'speechiness': [],
             'acousticness': [],
             'instrumentalness': [],
             'liveness': [],
             'valence': [],
             'tempo': [],
             'type': [],
             'id': [],
             'uri': [],
             'track_href': [],
             'analysis_url': [],
             'duration_ms': [],
             'time_signature': []
        }

    counter = 0
    
    for index, row in list_of_songs.iterrows():
        counter += 1
        print(f'downloading song {index}')
        song_id = search_song(row['title'], row['artist']) 
        features = sp.audio_features(song_id)[0]
        
        if features == None:
            continue
        
        d['title'].append(row['title'])
        d['artist'].append(row['artist'])
        d['danceability'].append(features['danceability'])
        d['energy'].append(features['energy'])
        d['key'].append(features['key'])
        d['loudness'].append(features['loudness'])
        d['mode'].append(features['mode'])
        d['speechiness'].append(features['speechiness'])
        d['acousticness'].append(features['acousticness'])
        d['instrumentalness'].append(features['instrumentalness'])
        d['liveness'].append(features['liveness'])
        d['valence'].append(features['valence'])
        d['tempo'].append(features['tempo'])
        d['type'].append(features['type'])
        d['id'].append(features['id'])
        d['uri'].append(features['uri'])
        d['track_href'].append(features['track_href'])
        d['analysis_url'].append(features['analysis_url'])
        d['duration_ms'].append(features['duration_ms'])
        d['time_signature'].append(features['time_signature'])
        
        if counter % 50 == 0:
            time.sleep(20)
        
    
    return pd.DataFrame(d)
        
        
```

```python
hotsongs = get_audio_features(data2)
nothotsongs = get_audio_features(data1)
```

```python
hotsongs.to_csv('hotsongs_with_features.csv', index = False)
```
